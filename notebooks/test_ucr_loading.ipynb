```python
import numpy as np

def load_ucr_numpy(file_path):
    # Load the data (skipping no rows, delimiter is tab)
    data = np.genfromtxt(file_path, delimiter='\t')
    
    # Split into labels and features
    y = data[:, 0].astype(int) # First column
    X = data[:, 1:]            # All other columns
    
    return X, y

# Usage:
# X_train, y_train = load_ucr_numpy("data/GunPoint/GunPoint_TRAIN.tsv")
```
