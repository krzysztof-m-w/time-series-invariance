```python
from notebooks.presets import set_cwd
set_cwd()
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dir_path = "data/simple"
for path_d in os.listdir(dir_path):
    for i, path in enumerate(
            os.listdir(os.path.join(dir_path, path_d))
        ):
        a = np.load(os.path.join(dir_path, path_d, path))
        plt.plot(a)
        plt.title(f"example {i} from {path_d}")
        plt.show()
        if i > 1:
            break
```
