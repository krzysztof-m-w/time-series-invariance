```python
from ts2vec import TS2Vec
import torch

model = TS2Vec(input_dims=14, device='cpu', output_dims=320)

model_path = 'training/ETTh1__my_test_run_20260218_110702/model.pkl'
model.load(model_path)

print("Model loaded successfully!")
```

```python
import os

import numpy as np

data_path = "../data/simple"
X  = []
labels = []
labels_names = []
names = [i for i in range(10)]

for i, dir_name in enumerate(os.listdir(data_path)):

    dir_path = os.path.join(data_path, dir_name)
    vectors = []
    for f in os.listdir(dir_path):
        path = os.path.join(dir_path, f)
        a = np.load(path)
        a = a[None, ..., None]
        a = np.repeat(a, 14, axis=2)
        v = model.encode(a, encoding_window='full_series')
        vectors.append(v.reshape(-1))

    X.extend(vectors)
    labels.extend([i] * len(vectors))
    labels_names.append(dir_name)
```

```python
X = np.array(X)
X.shape
```

```python
from sklearn.metrics import silhouette_score
sil = silhouette_score(X, labels, metric="cosine")
print("Silhouette:", sil)
```
