```python
from notebooks.presets import set_cwd
set_cwd()
import numpy as np
from src.embeddings.magnitude_spectrum import embedding_vector
import matplotlib.pyplot as plt
import os
```

```python
data_path = "data/simple"
X  = []
labels = []
labels_names = []
names: list[str]

for i, dir_name in enumerate(os.listdir(data_path)):

    dir_path = os.path.join(data_path, dir_name)
    vectors = []
    for f in os.listdir(dir_path):
        names = []
        path = os.path.join(dir_path, f)
        a = np.load(path)
        v, names = embedding_vector(a, 100, [(0, 25), (25, 50), (50, 75), (75, 100)])
        vectors.append(v)

    X.extend(vectors)
    labels.extend([i] * len(vectors))
    labels_names.append(dir_name)
    
```

```python
from sklearn.metrics import silhouette_score
sil = silhouette_score(X, labels, metric="cosine")
print("Silhouette:", sil)
```

```python
X = np.array(X)
X.shape
```

```python
group_size = 5

for i in range(X.shape[1]):
    column = X[:, i]

    groups = [column[j:j+group_size] 
              for j in range(0, len(column), group_size)]

    labels = [f"{labels_names[k]}" for k in range(len(groups))]

    plt.boxplot(groups, labels=labels)
    plt.title(names[i])
    plt.xticks(rotation=30)
    plt.show()
```
