```python
from notebooks.presets import set_cwd
set_cwd()
import matplotlib.pyplot as plt
import numpy as np
```

```python
def phase_space_image(ts, tau=64, m=4, bins=64):
    n = len(ts)
    valid = n - (m - 1) * tau
    X = np.column_stack([ts[i:i+valid] for i in range(0, m*tau, tau)])

    H, _, _ = np.histogram2d(X[:, 0], X[:, 1], bins=bins)
    H = H.astype(float)

    return H / (H.max() + 1e-12)   # avoid divide-by-zero
```

```python
def mutual_information(x, max_tau, bins=32):
    x = np.asarray(x)
    mi_values = []

    for tau in range(1, max_tau + 1):
        x1 = x[:-tau]
        x2 = x[tau:]

        # 2D histogram
        H, _, _ = np.histogram2d(x1, x2, bins=bins)
        Pxy = H / np.sum(H)

        # Marginals
        Px = np.sum(Pxy, axis=1)
        Py = np.sum(Pxy, axis=0)

        # Avoid log(0)
        nz = Pxy > 0
        mi = np.sum(Pxy[nz] * np.log(Pxy[nz] / (Px[:, None] * Py[None, :])[nz]))

        mi_values.append(mi)

    return np.array(mi_values)
```

```python
def show_plot(example_path):
    a = np.load(example_path)
    a_new = phase_space_image(a)

    plt.plot(a)
    plt.show()

    plt.figure(figsize=(5, 5))
    plt.imshow(a_new, cmap="binary", origin="lower")
    plt.title("Phase Space Image")
    plt.xlabel("Time")
    plt.ylabel("Time")
    plt.colorbar()
    plt.tight_layout()
    plt.show()
```
