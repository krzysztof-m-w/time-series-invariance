```python
from notebooks.presets import set_cwd
set_cwd()
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_plot(example_path):
    a = np.load(example_path)
    a = np.asarray(a).ravel()                          # ensure 1D
    a_new = np.tile(a, (len(a), 1))[np.newaxis, ...]   # shape -> (1, N, N)

    plt.plot(a)
    plt.show()

    plt.figure(figsize=(5, 5))
    plt.imshow(a_new[0], cmap="binary", origin="lower")
    plt.title("Recurrence Plot")
    plt.xlabel("Time")
    plt.ylabel("Time")
    plt.colorbar()
    plt.tight_layout()
    plt.show()
```

```python
example_path = "data/simple/constants_0/1.npy"
show_plot(example_path)
```
