```python
from notebooks.presets import set_cwd
set_cwd()
from pyts.image import MarkovTransitionField
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_plot(example_path):
    a = np.load(example_path)
    mtf = MarkovTransitionField(n_bins=32)
    a_new = mtf.fit_transform(a.reshape(1, -1))
    plt.plot(a)
    plt.show()

    plt.figure(figsize=(5, 5))
    plt.imshow(a_new[0], cmap="binary", origin="lower")
    plt.title("Markov Transition Field")
    plt.xlabel("Time")
    plt.ylabel("Time")
    plt.colorbar()
    plt.tight_layout()
    plt.show()
```

```python
example_path = "data/simple/constants_0/1.npy"
show_plot(example_path)
```
