```python
from notebooks.presets import set_cwd
set_cwd()
from scipy.signal import stft
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_plot(example_path):
    a = np.load(example_path)
    _, _, Zxx = stft(a, nperseg=128, noverlap=64)
    a_new = np.abs(Zxx)

    plt.plot(a)
    plt.show()

    plt.figure(figsize=(5, 5))
    plt.imshow(a_new, cmap="binary", origin="lower")
    plt.title("Recurrence Plot")
    plt.xlabel("Time")
    plt.ylabel("Time")
    plt.colorbar()
    plt.tight_layout()
    plt.show()
```

```python
example_path = "data/simple/constants_0/1.npy"
show_plot(example_path)
```
